```python
#This Python notebook will attempt to validate your CSV predictions file. It does this by checking
#that the file contains one prediction for all 300 trading days for which you were not provided the
#next day's closing price, and that each prediction is numerically reasonable. 
#This notebook does NOT check whether your predictions are correct!!

#INSTRUCTIONS:
#1. Upload your CSV predictions file to your Colab notebook's session storage.
#2. Type the name of the predictions file that you would like to validate below. 
#   For example: filename = 'Soper, Daniel.csv'
#3. Run the notebook!
filename = 'Connolly, Sean.csv'

#import the pandas library
import pandas as pd

#define list of acceptable day values
days = list(range(1001, 1301))

#load predictions from the CSV file
df = pd.read_csv(filename)

#declare a variable to hold the results
results = ''

#check column names
if 'day' not in df.columns:
  results += 'The CSV file does not contain a column named "day".\n'
elif 'next_day_close_price' not in df.columns:
  results += 'The CSV file does not contain a column named "next_day_close_price".\n'
else: #if both required columns exist
  #check values of the 'day' column
  for day in days:
    #if this day does not exist in the 'day' column
    if day not in df.day.values:
      results += 'The CSV file does not contain a prediction for day {}.\n'.format(day)
    else: #if this day exists in the CSV file
      #get the predicted value of 'next_day_close_price' for this day
      prediction = df[df.day == day].next_day_close_price.values[0]
      #if the prediction is not a number
      if not isinstance(prediction, (int, float)):
        try:
          prediction = float(prediction)
        except:
          results += 'The prediction for day {} is not a number.\n'.format(day)
      #if the prediction is a number
      if isinstance(prediction, (int, float)):
        #if the predicted value is unreasonable
        if prediction < 0:
          results += 'The prediction for day {} is negative, which is unreasonable for a stock price.\n'.format(day)

#if everything appears to be OK
if results == '':
  results = 'Your CSV file appears to be valid!'

#print the results
print(results)
```
